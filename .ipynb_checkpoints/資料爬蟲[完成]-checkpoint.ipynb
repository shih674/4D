```python
import requests
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from datetime import datetime
import json
import os
import time
datetime.now()
```

```python
yearList = [2020]
monthList = [1,2,3,4,5,6,7,8]

yearList0 = [109]
monthList0 = [1,2,3,4,5,6,7,8]


# 開啟視窗
url = 'https://www.twse.com.tw/zh/page/trading/fund/T86.html'
driver = webdriver.Chrome()
driver.get(url)

for ele0 in monthList:
    new_month = True
    for ele1 in range(1,32):
        print(datetime.now())
        stop = False
        month= ele0
        day_index = ele1
        
        # 選取月份
        if new_month == True:
            try:
                select1 = Select(driver.find_element_by_name('mm'))
                select1.select_by_value(str(month))  
                print(f':: 選取 {month}月份')
                new_month = False
            except:
                print("error in month")
                continue
        
        # 選取日期
        try:
            select3 = Select(driver.find_element_by_name('dd'))
            select3.select_by_index(day_index)
            print(f':: 選取 {day_index+1}日期')
        except:
            print("error in day")
            continue
        
        # 選取類別>>全部
        select2 = Select(driver.find_element_by_name('selectType'))
        select2.select_by_value("ALL")
        print(':: 選取類別 - 全部')
        
        # 按下查詢
        python_button = driver.find_element_by_class_name('main').find_element_by_css_selector('a')
        python_button.click()
        print(':: 按下查詢')     
        
        time.sleep(1)
        
        # 選取數量>>全部
        try:
            select4 = Select(driver.find_element_by_name('report-table_length'))
            select4.select_by_value("-1")
            print(':: 選取數量 - 全部')
        except:
            print(f'::{month}月{day_index}日 今天假日\n')
            print('==========================\n\n')
            continue

        time.sleep(1)

        # 按下下載
        python_button = driver.find_element_by_class_name('csv')
        python_button.click()
        print(f'::{month}月{day_index+1}日 下載完成\n')
        print('==========================\n\n')
        
        time.sleep(1)


# 關閉視窗
driver.close()
```
