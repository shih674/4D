```python
tw50 = [1101,1102,1216,1301,1303,1326,1402,2002,2105,2207,
        2301,2303,2308,2317,2327,2330,2357,2382,2395,2408,
        2412,2454,2474,2633,2801,2823,2881,2882,2883,2884,
        2885,2886,2887,2888,2890,2891,2892,2912,3008,3045,
        3711,4904,4938,5871,5876,5880,6505,6669,9910]

for j in range(0,len(tw50)):
    # 組裝廠
    for i in range(1,9):
        txt_name = str(tw50[j])+'_20200'+str(i)+'.txt'
        txt_path = os.path.join(folder_path, txt_name)
        print(f'檔案名稱 : {j},{txt_path}')
        with open(txt_path,'r',encoding='utf8') as f:
            txt_content = f.read()

        end_row = []
        for i,ele in enumerate(txt_content.split('\n')):
            #print(i,'\t',ele)
            if '各日成交資訊' in ele:
                start_row = i+3
            elif '說明' in ele:
                end_row.append(i) 
                #print(end_row,'==')
        #print(txt_content.split('\n')[start_row:end_row[0]])
        step1_content = txt_content.split('\n')[start_row:end_row[0]]

        step2_content = []
        for ele in step1_content:
            if ele =='':
                continue
            #print(ele)
            step2_content.append(ele)

        step3_content = []
        for i in range(len(step2_content)):
            temp = step2_content[i].split('",')[:-1]
            #print(temp)
            for i in range(len(temp)):
                temp[i] = temp[i].replace('"','').replace(',','')
            print(temp)
            step3_content.append(temp)

        df = pandas.DataFrame(columns=["日期","成交股數","成交金額","開盤價","最高價","最低價","收盤價","漲跌價差","成交筆數",],
                              data = step3_content)

        csv_path = txt_path.replace('.txt','.csv')
        csv_path

        df.to_csv(csv_path, index=False,encoding='utf8')
        print(f'儲存檔案: {csv_path}\n\n=======================================\n')
```
